```python
import tensorflow as tf
```

```python
with tf.Graph().as_default():
    primes = tf.constant([1,2,3,4,5,6],dtype=tf.int32)
    ones = tf.ones([6],dtype=tf.int32)
    add_vec = tf.add(primes,ones)
    with tf.Session() as sess:
        print(add_vec.eval())
```

```python
with tf.Graph().as_default():
    scalr = tf.zeros([])
    vector = tf.ones([3])
    matri = tf.zeros([2,3])
    with tf.Session() as sess:
        print('sclar has shape {} and values as {}'.format(tf.shape(scalr),scalr.eval()))
        print('vector has shape {} and values as {}'.format(tf.shape(vector),vector.eval()))
        print('matri has shape {} and values as {}'.format(tf.shape(matri),matri.eval()))
```
