```python
import tensorflow as tf
```

```python
with tf.Graph().as_default():
    x = tf.constant(5, name="x_value")
    y = tf.constant(7, name="y_value")
    x_y_add = tf.add(x,y,name="x_y_add")
    with tf.Session() as sess:
        x_y_add.eval()
        writer = tf.summary.FileWriter('/Users/rohishah/log_file',sess.graph)
        writer.add_graph(tf.get_default_graph())
```
